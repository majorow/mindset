```python
# Подключаем необходимые библиотеки
from ultralytics import YOLO
import numpy as np
import pandas as pd
```

```python
# Создание новой модели
path = '/content/drive/MyDrive/mindset/dataset/'
model = YOLO('yolov8n.pt')
model.train(data=path + 'data.yaml', epochs=20)
```

###Проверка результата

```python
model = YOLO('/content/drive/MyDrive/mindset/dataset/best.pt')
res = model.predict('/content/drive/MyDrive/mindset/testA/1.jpg')
np_res = res[0].boxes.cpu().boxes.numpy()
print('Количество окон -', len(np_res))
```
